```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rewards = np.loadtxt('rewards.txt')
```

```python
f, ax = plt.subplots(figsize=(7,5))

ax.plot(np.arange(len(rewards)) + 1, rewards, 'k.--', linewidth=0.3)
ax.set_xlabel('Episode', size=15)
ax.set_ylabel('Reward', size=15)

# ax.set_xticks(np.arange(len(rewards))+1)
ax.set_yticks([-3, -2, -1, 0, 1])
plt.show()
```
